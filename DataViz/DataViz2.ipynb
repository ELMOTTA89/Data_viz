## Visualize Data

### <font color='purple'> Quelle est la répartition des passagers par classe ? </font>
-> Single categorical variable (une seule variable qualitative nominale)

```python
repartition=df_titanic["Pclass"].value_counts()
repartition
```

```python
sns.countplot(x='Pclass',data=df_titanic)
plt.title('repartition passagers par class')
plt.xlabel('class')
plt.ylabel('Nombre de passagers')
plt.show()
```

### <font color='purple'> Quelle est la répartition des passagers par sexe ? </font>
-> type of variable(s)?

```python
sns.countplot(x='Sex',data=df_titanic)
plt.title('repartition passagers par class')
plt.xlabel('class')
plt.ylabel('Nombre de passagers')
plt.show()
```

### <font color='purple'>Distribution des passagers selon âge et sexe </font>
-> type of variable(s)?

```python
g=sns.FacetGrid(df_titanic,col="Sex",height=5,aspect=1.2)
g.map(sns.histplot,"Age",bins=20,kde=False)
g.set_axis_labels("Age","Nombre de passagers")
g.set_titles("distribution des ages")
```

### <font color='purple'>Quel est le taux de survie selon le sexe ? l’âge ? la classe ? </font>
-> type of variable(s)?

```python
survie_par_sexe=df_titanic.groupby('Sex')['Survived'].mean()
```

```python
survie_par_class=df_titanic.groupby('Pclass')['Survived'].mean()
```

```python
sns.barplot(x=survie_par_sexe.index,y=survie_par_sexe.values,palette="pastel")
plt.title('Taux de survie selon le sexe')
plt.xlabel('sexe')
plt.ylabel('Taux de surviue')
plt.show()
```

```python
sns.barplot(x=survie_par_class.index,y=survie_par_class.values,palette="pastel")
plt.title('Taux de survie selon le sexe')
plt.xlabel('sexe')
plt.ylabel('Taux de surviue')
plt.show()
```

### <font color='purple'> Y a-t-il un lien entre prix payé pour le billet et l’âge des passagers ? </font>
-> type of variable(s)?

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='Age',y='Fare',data=df_titanic,hue='Pclass',palette='deep')
plt.title('relation entre lage et le prix de billet')
plt.xlabel('Age')
plt.ylabel('Prix de billet')
```
